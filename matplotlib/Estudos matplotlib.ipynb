# Estudos Matplotlib - Uma poderosa biblioteca para construção de gráficos.

```python
# Definição da subclasse pyplot
import matplotlib.pyplot as plt
```

```python
#Linha de comando para conseguir ver os gráficos no jupyter
%matplotlib inline 
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x*x
```

```python
y
```

```python
# Funcional
plt.plot(x,y,'ro')
plt.xlabel('eixo X')
plt.ylabel('eixo Y')
plt.title('Gráfico X x Y')
plt.show()
```

# Possibilidade de criar mais de um gráfico na mesma imagem, para isso vamos utilizar a função subplot.

```python
# Primeiro parametro: Número de linhas.
# Segundo parametro: Número de colunas.
# Terceiro parametro: O gráfico que vou acessar no momento.
plt.subplot(1,2,1)
plt.plot(x,y,'ro-')
plt.subplot(1,2,2)
plt.plot(y,x,'go-')
```

# Esta é a forma mais comum de criar gráficos, porém este caminho não dá acesso a todas as possibilidades.

```python
# Cria uma figura vazia, caixa preta
fig = plt.figure()
# Definição de eixos com 4 parametros.
# Os dois primeiros representam a posição considerando o ponto inferior esquerdo da imagem.
# Os dois últimos representam largura e altura.
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.15,0.55,0.3,0.3])
axes1.plot(x,y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Titulo')

axes2.plot(y,x)
```

# Continuação da aula

```python
# Utilização de geração de plots a partir de objetos da subclasse plt
fig, ax = plt.subplots()
ax.plot(x,x**3,'b--')
ax.plot(x,x**4,'g--')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Gráfixo X x Y')
```

```python
# Manipulação de subplots pela metodologia de instancia de objetos
fig,ax = plt.subplots(nrows=1,ncols=2)
```

```python
i=1
for axe in ax:
    axe.plot(x,y**2 + (i/2)*y + i,'b')
    axe.set_title('Titulo ' + str(i))
    i += 1
fig
```

```python
fig,ax = plt.subplots()
ax.plot(x,y,'r-',label='Dados 1')
ax.plot(y,x,'g-',label='Dados 2')
ax.legend()
```
