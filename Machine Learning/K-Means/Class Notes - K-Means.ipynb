```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Import make_blobs to generate data for K-Means Studies
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2,centers=4,cluster_std=1.8,random_state=50)
```

```python
# This method return a tuple with data and the cluster that each point is associate.
# The first position tuple[0] --> return data
# The second position tuple[1] --> return the associate cluster
type(data)
```

```python
# I'll Create a scatter plot just to see the segregation
plt.figure(figsize=(12,8))
plt.scatter(data[0][:,0], data[0][:,1],cmap='rainbow',c=data[1])
```

It's possible to see 4 clusters!

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f,(ax1,ax2) = plt.subplots(1,2,sharey=True, figsize=(12,8))
ax1.set_title('Kmeans')
ax1.scatter(data[0][:,0], data[0][:,1],cmap='rainbow',c=kmeans.labels_)
ax2.set_title('Real Data')
ax2.scatter(data[0][:,0], data[0][:,1],cmap='rainbow',c=data[1])
```
