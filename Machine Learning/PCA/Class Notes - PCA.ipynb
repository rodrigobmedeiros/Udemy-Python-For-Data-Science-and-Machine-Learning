```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
df = pd.DataFrame(cancer['data'],columns = cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
# Now, the variance became unity
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
# parameter n_components define the number of principal components we want to use.
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
x_pca.shape
```

```python
scaled_data.shape
```

We can see that using the PCA method I can reduce the space of features that will be analyzed to predict things.

```python
plt.figure(figsize=(12,8))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```
